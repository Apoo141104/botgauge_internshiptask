**1.Function Summarization:Defining all the functions and their purpose, functionality and dependencies on other libraries, functions.**

**signup(username, password):**
***Purpose***: Registers a new user by storing a hashed password in the MySQL database.
***Functionality***: It takes the username and password, hashes the password using SHA-256, and inserts the data into the users table in the MySQL database.
**Dependencies**: Depends on hashlib for hashing and mysql.connector for database interaction.

```python
# Function to generate AST for a given code snippet
def generate_ast(function_code):
    tree = ast.parse(function_code)
    return ast.dump(tree, indent=4)
signup_code = '''
def signup(username, password):
    hashed_password = hashlib.sha256(password.encode()).hexdigest()
    query = "INSERT INTO users (username, password) VALUES (%s, %s)"
    values = (username, hashed_password)
    cursor.execute(query, values)
    db.commit()
'''

print(generate_ast(signup_code))
```

**signin(username, password)**:
**Purpose**: Authenticates a user by verifying the hashed password.
**Functionality**: It checks if the given username and password match an entry in the users table of the database.
**Dependencies**: Similar to signup, uses hashlib and mysql.connector.

```python
signin_code = '''
def signin(username, password):
    hashed_password = hashlib.sha256(password.encode()).hexdigest()
    query = "SELECT * FROM users WHERE username = %s AND password = %s"
    values = (username, hashed_password)
    cursor.execute(query, values)
    user = cursor.fetchone()
    return user is not None
'''

print(generate_ast(signin_code))
```

**get_weather**(city):
**Purpose**: Fetches the current weather for a specified city using the OpenWeatherMap API.
**Functionality**: Sends an HTTP GET request to the weather API and retrieves weather information (temperature and description).
**Dependencies**: Uses the requests library for making API calls.


```python
get_weather_code = '''
def get_weather(city):
    weather_url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={weather_api_key}"
    response = requests.get(weather_url)
    data = response.json()
    if data["cod"] == 200:
        temperature = data["main"]["temp"] - 273.15  # Convert to Celsius
        weather_description = data["weather"][0]["description"]
        return f"Temperature in {city}: {temperature:.2f}째C, {weather_description.capitalize()}"
    else:
        return "Weather information not available."
'''

print(generate_ast(get_weather_code))
```

**Functionality**: Sends an HTTP GET request and parses the response to display details about the latest earthquake.
**Dependencies**: Uses the requests library for API calls.

```python

get_earthquake_info = '''
def get_earthquake_info():
    response = requests.get(earthquake_api_url)
    data = response.json()
    features = data.get("features")
    if features:
        earthquake = features[0]
        properties = earthquake.get("properties")
        if properties:
            magnitude = properties.get("mag")
            place = properties.get("place")
            if magnitude is not None and place is not None:
                return f"Earthquake Alert!\\nMagnitude: {magnitude}\\nLocation: {place}"
            else:
                return "Earthquake data is incomplete."
        else:
            return "Earthquake properties not found."
    else:
        return "No recent earthquakes."
'''

print(generate_ast(get_earthquake_info))
```

**2.Search Mechanism :To Find keywords like function name, variable name in the code .To find variables according to data type in the code.**

```python
def search_keyword_in_code(code, keyword):
    lines = code.split('\n')
    # Using enumerate to get the line number along with the line content
    return [(index + 1, line) for index, line in enumerate(lines) if keyword in line]

code_snippet = main_content
x=input("Enter the keyword such as function name, variable name : ")
search_results = search_keyword_in_code(code_snippet, x)

print("Occurrences of ",x,'(line number,line content):', search_results)
```

```python
import ast

# Function to search for specific data types (like integers, strings, lists) in code
def search_data_type_in_code(code, data_type):
    tree = ast.parse(code)
    occurrences = []

    # Map Python data types to AST nodes
    data_type_mapping = {
        'int': ast.Constant,   # 'ast.Constant' covers numbers (int, float) and strings in Python 3.8+
        'float': ast.Constant,
        'str': ast.Constant,
        'list': ast.List,
        'dict': ast.Dict,
        'tuple': ast.Tuple
    }

    # Traverse the AST and check for the specified data type
    for node in ast.walk(tree):
        if isinstance(node, ast.Assign):
            # Check if the assigned value matches the data type
            for target in node.targets:
                if isinstance(target, ast.Name):
                    variable_name = target.id
                    if isinstance(node.value, data_type_mapping[data_type]):
                        # For 'Constant', we need to check the exact type of the value
                        if data_type == 'int' and isinstance(node.value.value, int):
                            occurrences.append((variable_name, node.lineno, 'int', node.value.value))
                        elif data_type == 'float' and isinstance(node.value.value, float):
                            occurrences.append((variable_name, node.lineno, 'float', node.value.value))
                        elif data_type == 'str' and isinstance(node.value.value, str):
                            occurrences.append((variable_name, node.lineno, 'str', node.value.value))
                        elif data_type in ['list', 'dict', 'tuple']:
                            occurrences.append((variable_name, node.lineno, data_type, node.value))
                    else:
                        # If the value is not a simple constant, it could be a more complex expression
                        if data_type in ['list', 'dict', 'tuple'] and isinstance(node.value, data_type_mapping[data_type]):
                            occurrences.append((variable_name, node.lineno, data_type, node.value))

        elif isinstance(node, data_type_mapping[data_type]):
            # For 'Constant', we need to check the exact type of the value
            if data_type == 'int' and isinstance(node.value, int):
                occurrences.append(("", node.lineno, 'int', node.value))
            elif data_type == 'float' and isinstance(node.value, float):
                occurrences.append(("", node.lineno, 'float', node.value))
            elif data_type == 'str' and isinstance(node.value, str):
                occurrences.append(("", node.lineno, 'str', node.value))
            elif data_type in ['list', 'dict', 'tuple']:
                occurrences.append(("", node.lineno, data_type, node))

    return occurrences

code_snippet = main_content
data_type_input = input("Enter the data type: ")
data_occurrences = search_data_type_in_code(code_snippet, data_type_input)
print("Occurrences of ", data_type_input, '(variable name,line number, datatype, value):', data_occurrences)
```
