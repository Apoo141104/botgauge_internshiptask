**loading dataset BeIR/fiqa**

```python
from datasets import load_dataset

corpus= load_dataset("BeIR/fiqa", "corpus")

queries= load_dataset("BeIR/fiqa", "queries")

# Display structure
print(queries)
print(corpus)


```

**Extracting text from the queries and corpus datasets**

```python
# Extract 'text' from the queries and corpus datasets directly
queries_list = [query['text'] for query in queries['queries']]
corpus_list = [corpus['text'] for corpus in corpus['corpus']]
```

importing the small embedding model -**'sentence-transformers/all-MiniLM-L6-v2'** from hugging face

```python
from sentence_transformers import SentenceTransformer, util
small_embedding_model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
```

encoding the text into **embeddings** through **small embedding model** and finding the **top 10 passages** for the **first query** and printing them

```python
# Encode the queries and corpus using the small model (MiniLM)
query_embeddings_small = small_embedding_model.encode(queries_list, convert_to_tensor=True)
corpus_embeddings_small = small_embedding_model.encode(corpus_list, convert_to_tensor=True)

# Retrieve top-k candidate passages using cosine similarity
top_k = 10  # Number of candidates
hits = util.semantic_search(query_embeddings_small, corpus_embeddings_small, top_k=top_k)

# Show retrieval results for a sample query
print(f"Top {top_k} hits for the first query: {hits[0]}")

```

```python
large_embedding_model = SentenceTransformer('Snowflake/snowflake-arctic-embed-l')
```

encoding the text into **embeddings** through **large embedding model** and finding the **top 10 passages** for the **first query** and printing them

```python
query_embeddings_large = large_embedding_model.encode(queries_list, convert_to_tensor=True)
corpus_embeddings_large = large_embedding_model.encode(corpus_list, convert_to_tensor=True)

# Retrieve top-k candidate passages using cosine similarity
top_k = 10  # Number of candidates
hits2 = util.semantic_search(query_embeddings_large, corpus_embeddings_large, top_k=top_k)

# Show retrieval results for a sample query
print(f"Top {top_k} hits for the first query: {hits2[0]}")
```

loading **1st ranking model** and **reranking the top-k passages** and providing reranking for the first query according to the small embedding model.

```python
# Check the content and structure of query_text and passages
print(f"Query Text: {query_text}")
print(f"Passages: {passages}")
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer

# Load ranking model
ranking_model = AutoModelForSequenceClassification.from_pretrained('cross-encoder/ms-marco-MiniLM-L-12-v2')
tokenizer = AutoTokenizer.from_pretrained('cross-encoder/ms-marco-MiniLM-L-12-v2')

# Rerank the top-k passages
def rerank(query, passages):
    inputs = tokenizer([query] * len(passages), passages, truncation=True, padding=True, return_tensors="pt")
    with torch.no_grad():
        outputs = ranking_model(**inputs)
    return outputs.logits.squeeze().tolist()

# Example of reranking for the first query
query_text = queries['queries'][0]['text']
passages = [corpus['corpus'][hit['corpus_id']]['text'] for hit in hits[0]]

reranked_scores = rerank(query_text, passages)
# Rank passages based on scores
sorted_passages = [passages[idx] for idx in torch.argsort(torch.tensor(reranked_scores), descending=True)]
print(f"Reranked passages for the query '{query_text}': {sorted_passages}")
```

**Benchmarking and Evaluation**

```python
import numpy as np

# Function to calculate DCG at rank k
def dcg_at_k(relevance_scores, k=10):
    relevance_scores = np.asfarray(relevance_scores)[:k]
    if relevance_scores.size == 0:
        return 0.0
    return np.sum((2 ** relevance_scores - 1) / np.log2(np.arange(1, relevance_scores.size + 1) + 1))

# Function to calculate NDCG@k
def ndcg_at_k(ground_truth_relevance, predicted_ranking, k=10):
    """
    Calculate NDCG at rank k
    Args:
        ground_truth_relevance: List of ground truth relevance scores
        predicted_ranking: List of predicted relevance scores (ranked)
        k: cutoff rank, default is 10
    """
    ideal_ranking = sorted(ground_truth_relevance, reverse=True)  # Ideal ranking
    dcg = dcg_at_k(predicted_ranking, k)
    idcg = dcg_at_k(ideal_ranking, k)

    return dcg / idcg if idcg > 0 else 0.0
```

```python
def compare_ndcg(ground_truth, predicted_without_ranking, predicted_with_ranking, k=10):
    """
    Compare NDCG@10 for two scenarios: without and with ranking models.

    Args:
        ground_truth: Ground truth relevance scores
        predicted_without_ranking: List of relevance scores without ranking models
        predicted_with_ranking: List of relevance scores with ranking models
        k: Rank cutoff for NDCG@k
    """
    ndcg_without_ranking = ndcg_at_k(ground_truth, predicted_without_ranking, k)
    ndcg_with_ranking = ndcg_at_k(ground_truth, predicted_with_ranking, k)

    print(f"NDCG@{k} without ranking: {ndcg_without_ranking:.4f}")
    print(f"NDCG@{k} with ranking: {ndcg_with_ranking:.4f}")

    return ndcg_without_ranking, ndcg_with_ranking
```

```python
def analyze_model_impact(ground_truth, model_combinations, k=10):
    """
    Analyze the impact of different model combinations on NDCG@10.

    Args:
        ground_truth: Ground truth relevance scores
        model_combinations: A list of dictionaries where each dictionary contains
                            'name': Model name,
                            'predicted': List of predicted relevance scores
        k: Rank cutoff for NDCG@k
    """
    results = []

    for model in model_combinations:
        model_name = model['name']
        predicted_ranking = model['predicted']
        ndcg_score = ndcg_at_k(ground_truth, predicted_ranking, k)
        results.append({'Model': model_name, 'NDCG@10': ndcg_score})
        print(f"{model_name} - NDCG@{k}: {ndcg_score:.4f}")

    return results
```
